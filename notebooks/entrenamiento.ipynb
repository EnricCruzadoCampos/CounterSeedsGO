```python
from ultralytics import YOLO

# Cargar el modelo base pre-entrenado
model = YOLO('yolov8n.pt')

# Entrenar
# data: apunta al archivo data.yaml que se descargó en la Celda 4
# epochs: cuántas veces repasa las fotos (50-100 es un buen número)
# imgsz: tamaño de las imágenes (640 es el estándar)
results = model.train(data='/content/dataset/data.yaml', epochs=50, imgsz=640, name='trigo_model')
```

```python
metrics = model.val()
print(f"Precisión (mAP50): {metrics.box.map50}")
```
