```python

# --- CELDA 3: ENTRENAR EL MODELO (VERSI√ìN YOLO11) ---
from ultralytics import YOLO

# 1. Cargar el modelo base YOLO11
# CAMBIO: 'yolo11n.pt' es la versi√≥n Nano (m√°s r√°pida) de la generaci√≥n 11.
# Si quieres m√°s precisi√≥n (pero un pel√≠n m√°s lento), usa 'yolo11s.pt'
model = YOLO('yolov8s.pt')

# 2. Entrenar
print(f"üöÄ Comenzando entrenamiento con YOLO11 en: {dataset.location}")

model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=100,
    imgsz=640,
    device=0,      # <--- IMPORTANTE: Fuerza el uso de la GPU
    plots=True     # Genera gr√°ficas de c√≥mo aprende
)
```
